```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
sns.set_style("darkgrid")
mpl.rcParams['figure.figsize'] = (20,5)



```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
neighbourhood_group = df['neighbourhood_group']
price = df['price']

plt.scatter(neighbourhood_group, price, s=75)

plt.xlabel('Price')
plt.ylabel('Neighbourhood Group')
plt.title('Correlation between Price and Neighbourhood Group in NYC')
plt.xticks(rotation=45)
plt.show()
```

```python

room_type_counts = df['room_type'].value_counts()
labels = room_type_counts.index
sizes = room_type_counts.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

plt.title('Room Type Distribution')
plt.axis('equal')
plt.show()
```

```python
prices = df['price']

plt.hist(prices, bins=80, edgecolor='black') 
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.show()
```

```python

sns.violinplot(x="neighbourhood_group", y="price", hue="room_type", data=df, inner="quartile")
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Price Distribution by Neighbourhood Group and Room Type')

plt.legend(title='Room Type')
plt.show()
```

```python
numerical_features = df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']]
correlation_matrix = numerical_features.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Price Correlation')
plt.show()
```

```python

sns.boxplot(x='neighbourhood_group', y='price', data=df)

plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Box Plot of Prices by Neighbourhood Group')


# Show plot
plt.show()
```

```python
df_sorted = df.sort_values(by='number_of_reviews', ascending=False)

plt.plot(df_sorted['number_of_reviews'], marker='o', color='b', linestyle='-')

# Add labels and title
plt.xlabel('Listing Index')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews for Each Listing')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
